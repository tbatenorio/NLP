```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

#print(fetch_20newsgroups.__doc__)
```

# **Importação dos Dados para Análise**

```python
raw_documents= fetch_20newsgroups(subset='train', return_X_y=True)[0]
N = 5

#Visuzaliação de alguns exemplos dos Documentos
for idx, doc in enumerate(raw_documents[:N]):
  print(f"========== Doc {idx} ==========")
  print(doc)
```

# **Tokenização de texto**

```python
# Criar um vetorizador de contagem (para calcular o TF)
count_vectorizer = CountVectorizer(stop_words='english', token_pattern=r"\b[A-Za-z]{3,}\b")
# Aprenda o dicionário de vocabulário e retorna a matriz de termos do documento
term_counts = count_vectorizer.fit_transform(raw_documents)
```

# **Cálculo dos Indicadores e Criação do DataFrame**

```python
# Criar um transformador IDF
tfidf_transformer = TfidfTransformer()
tfidf_transformer.fit(term_counts)

# Calcular TF, IDF e TF-IDF para todos os termos
tf = count_vectorizer.transform(raw_documents)
idf = tfidf_transformer.idf_
tfidf = tfidf_transformer.transform(tf)

# Obter os termos do vetorizador
terms = count_vectorizer.get_feature_names_out()

# Criar um DataFrame com os resultados
df = pd.DataFrame(data={'Termo': terms, 'TF': tf.sum(axis=0).A1, 'IDF': idf, 'TF-IDF': tfidf.sum(axis=0).A1})

df
```

# **Análise do TOP-5 Tokens p/ Indicador**

```python
df.sort_values(by='TF')
```

```python
df.sort_values(by='IDF')
```

```python
df.sort_values(by='TF-IDF').tail(10)
```

**Conclusão:**

Temos que quando maior o valor do indicador **TF-IDF**, maior é a relevância de um terminado **Termo/Token** no conjunto de documentos. Tendo isso em vista, obtivemos o seguinte TOP-10 Tokens:

**TOP 5:**

1.   edu
2.   com
3.   subject
4.   lines
5.   organization

Termos como **'edu'**, **'com'** e **'subject'** estão associados principalmente a endereços e cabeçalhos de E-mail, visto que nossa coletânia de Documentos se trata de E-mails aparentemente de âmbito universitário.

Os Termo **'lines'** e **'organization'** aparecem em todos os E-mails ao menos 1x, especificando o numero de linhas no corpo do e-mail e a organização de origem do documento.

Por serem E-mails, o conteúdo dos Documentos tendem a ter um conjunto de palavras relativamente pequeno, o que influencia nos cálculos dos Indicadores.

**Top 6-10:**

6. writes
7. article
8. university
9. posting
10. don

Seguindo o comportamento descrito anteriormente, o termo **'posting'**, **'university'** pode ser encontrado em boa partes dos cabeçalhos.

Os termos **'article'** e **'writes'** dão a conotação de que a nossa base de dados são uma série de E-mails que tratam de *comentários, indagações e discussões* a cerca de artigos científicos.

O termo **'don'** deve ser disconsiderado, pois não tem relevância semântica para o conjunto de documentos.
